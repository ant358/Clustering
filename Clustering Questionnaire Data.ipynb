## The Data  
The data comprises the answers of 206 subjects to 32 questions. The questions were multiple choice selections with responses ranging from 1 - 5. The data is not labelled, i.e. we do not know what the questions were or what the answers to the questions mean.

## The Research Objectives  
There are three primary objectives to the analysis of this data:  
    1) To group the subjects by their answers  
    2) Can the dimensionality of the data be reduced - are there redundant or nested questions?  
    3) This analysis was blind. Does the result match the traditional analysis and groupings?

Unfortunately, the final result to see if the clustering matches the traditional analysis can not be shown here as it is confidential!

## Exploration of the provided data 
### Load and Inspect the data  
The data was provided in a csv file - it will be loaded into a pandas dataframe for these analyses.

```python
# import the basic python modules required
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from scipy.cluster.hierarchy import linkage, dendrogram
from scipy.cluster.hierarchy import fcluster
```

```python
# import the data from the csv file provided
q_data = pd.read_csv('./questionnaire_data.csv')
```

```python
# set the index to subject id
q_data.set_index('Subject_id', inplace=True)
```

```python
# check the dataframe created, does it looks right?
print('Table 1.0. The first 5 rows of the original data provided.'
      ' Loaded into a dataframe and indexed by the subject Id\'s\n ')
q_data.head()
```

```python
# check the number of rows and columns
print('There are {} subject Id\'s and {} questions'.format(
    q_data.shape[0], q_data.shape[1]))
```

### What are the distributions of the answers to each question?  
Does it look like the questions were answered randomly?

```python
# Loop through the question answers and produce a histogram of the answers
cols = q_data.columns
fig = plt.figure(figsize=(12, 12))
for i, c in enumerate(cols):
    ax = fig.add_subplot(8, 4, i+1)
    q_data[c].value_counts(sort=False).plot(kind='bar', ax=ax)
    ax.set_title(c + " Answer Distribution", color='red')
    plt.ylim(0, 150)
    # plt.grid(b=None)
print('Fig. 1.0 The distribution of the answers to the questions\n ')
fig.tight_layout()
plt.show()
```

### Are the answers to the questions correlated with each other?

```python
# generate a correlation matrix
corr = q_data.corr()

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, cbar_kws={"shrink": .5})

print('Fig.1.1 The correlations of the questions and'
      ' answers to each other.\n Displayed in a heatmap.\n ')
plt.show()
```

### What were the questions with the highest correlations?
From the heatmap it can be seen that some questions have a high positive correlation. There a lower negative correlation between the questions. Which were the most correlated questions?

```python
# make all the correlations positive
corr = corr.abs()

# replace the 100% cross correlations were the value is 1 change to zero
corr.replace(to_replace=1, value=0, inplace=True)

# stack the columns
corr = corr.stack()

# sort the values
corr = corr.sort_values(ascending=False)
```

```python
# what were the top 10 correlations
print('Table 1.1 The top 10 correlations'
      '(20 including reverse correlations)\n')
corr[:20]
```

### Conclusion of the initial data exploration

From the answer distributions in Fig.1 and the 0.70 correlations shown in Fig 1.1 heatmap and Table 1.1 it is clear without doing any more statistical analysis that these questions were not answered randomly and that there is a structure and relationship between the questions and answers.

## Data Analysis
### Principle Component Analysis (PCA)  
PCA can be used to illustrate how many questions in the data set are required to explain the variance in the subjects answers. Can the data set be explained by using a smaller subset of the questions?

```python
# create and instance - without specifying the number of components required
pca = PCA()

# fit the data
pca.fit(q_data)

# create a dataframe to store the outputs of the PCA analysis
# set the index to the question numbers
q_data_output = pd.DataFrame(index=cols)

# collect the explained variance attribute
q_data_output['variance_ratio'] = pca.explained_variance_ratio_

# cumulatively sum the explained variance of each question
q_data_output['cum_var'] = q_data_output['variance_ratio'].cumsum(axis=0)
```

```python
fig = plt.figure(figsize=(8, 5))
# plot the data and add a line to indicate 95%
q_data_output['cum_var'].plot()
plt.hlines(0.95, q_data_output.index[0], len(q_data.columns), 'r', linestyles='dashed')
plt.hlines(0.5, q_data_output.index[0], len(q_data.columns),
           'orange', linestyles='dashed')
plt.ylabel('% of the explained variance')
plt.title('The cumulative number of questions required'
          ' \nto explain the variance in the dataset')

print('Fig 2.0 \n')
plt.show()
```

```python
# how many questions explain 50% of the data?
pca2 = PCA(n_components=0.5)
pca2.fit(q_data)
print('The number of questions required to explain 50% '
      'of the variance in this data set: {}'.format(
          pca2.n_components_))
```

```python
# how many questions explain 95% of the data?
pca3 = PCA(n_components=0.95)
pca3.fit(q_data)
print('The number of questions required to explain '
      '95% of the variance in this data set: {}'.format(
          pca3.n_components_))
```

#### Summary of the PCA Results  
In Fig.2.0, you can see that there is an initial steep rise with up to 5 of the questions explaining 50% of the data's overall variance. It takes 26 questions to explain 95% of the variance in the data. There is no clear 'elbow' observed in the graph. In conclusion, there is little scope for dimensionality reduction in this dataset, and nearly all of the questions are important and required to explain the results.

### Clustering  
The questionnaire was used to classify the subjects into distinct groups. Can the groups be derived from the data without knowing the context and structure of the questions? 
How does this unsupervised clustering compare to the original analysis? 

#### K-Means clustering  
First we will try and use K-Means clustering algorithm, to visualise the multi dimensional data in a 2D space and see how the data clusters. We can use the K-Means attribute inertia to measure the quality of the cluster separation. A plot of cluster inertia vs the number of clusters can be used to determine an ideal number of clusters for the data.

```python
# create a list of cluster numbers
clusters = list(range(1, 21))
#create a list to hold the inertia values
inertia = []
# loop through clusters and fit the data to each cluster size and collect the inertia data
for c in clusters:
    # create an instance
    kmn = KMeans(n_clusters=c)
    kmn.fit(q_data)
    inertia.append(kmn.inertia_)
    
df = pd.DataFrame({'inertia' : inertia}, index=clusters) 
df.plot(marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.title('The Reduction in K-Means inertia \nwith an increasing number of clusters')
plt.xlim(0,20)
plt.xticks(clusters)
print('Fig 2.1 \n')
plt.show()
```

#### K-Means Inertia Results  
The graph in Fig 2.1 shows that, similar to the PCA analysis above, there is not a distinct 'elbow' in the graph and therefore it does not suggest an optimum number of clusters to use.

#### Plot the K-Means  
How does the data look when clustered with K-Means? Below we plot the data projected into 2 dimensions, with 8 clusters to visualise how the well the clusters separate. 

```python
# generate and 8 clusters
kmn = KMeans(n_clusters=8, random_state=42)
kmn.fit_predict(q_data)

# get the cluster labels
Cluster = kmn.labels_
```

```python
# use PCA to convert the 32 questions into a 2D space to graph the clusters
pca = PCA(n_components=2)
X = pca.fit_transform(q_data)
```

```python
# create a dataframe to hold the data to plot
km_df = pd.DataFrame({'x' : X[:,0], 'y' : X[:,1], 'Cluster' : Cluster}, index=q_data.index)
```

```python
# plot the clusters
plt.figure(figsize=[12,7.5])
sns.scatterplot(data=km_df, x='x', y='y', hue='Cluster', palette='rainbow', legend='full', s=150)
#plt.scatter(km_df['x'], km_df['y'], c=km_df['Cluster'], cmap='rainbow', s=150)
plt.title('K-Means clustering with 206 subjects grouped into 8 clusters \nand projected onto a 2 dimensional representation of the questionnaire data set')

print('Fig 2.2 \n')
plt.show()
```

```python
# How many subjects are there in each cluster?
print('Table 2.1 The number of subjects in each of the 8 K-Means clusters\n ')
print(km_df['Cluster'].value_counts(sort=False))
```

#### Summary of the K-Means Clustering results 
Fig.2.2 illustrates why the inertia shown in Fig.2.1 did not show an optimum number of clusters. It can be seen that clusters 0, 7, 3 are diverse and lay outside the main core of the data. The other clusters (1, 2, 4, 5 & 6)  lay close together in this 2D projection of the data and they are hard to separate.

#### Hierarchical Clustering with a dendogram  
Given that these types of subject surveys typically try to group and sub group the subjects by their characteristics; a hierarchical clustering approach may produce a better way of understanding the data. To do this a linkage function can be used to measure the distance between the observations in the data set and form the clusters. Starting from the nearest two points these are then linked to form a single point and the distance to the next point evaluated. The process repeats until you have a tree like structure. However there are several algorithms for computing the linkages and the statistical literature does not provide any clear direction on which to use. First we will try them all and see which methods produce a realistic clustering of the data.

```python
methods = ['median', 'single', 'average', 'weighted', 'centroid', 'complete', 'ward']
fig=plt.figure(figsize=(15,15))
for i, c in enumerate(methods):
    ax=fig.add_subplot(4,2,i+1)
    mergings = linkage(q_data, method=c) 
    dendrogram(mergings,
           labels=None,
           orientation='right', 
           leaf_rotation=0 )
    ax.set_title(c +" distance measure",color='k')
print('Fig. 2.4 The clustering produced by different distance methods\n ')
fig.tight_layout()
plt.show()
```

#### Hierarchical Clustering with a dendogram using the ward distance measure

From Fig 2.4 it can be seen that when either the ward or complete method is used to measure the distance between data points a converging dendogram is produced as we would expect from this type of survey. The ward method was selected as it appears to better generalise the data into even clusters. A detailed dendogram is produced below, using a distance of 15 the data can be clustered into 6 characteristic groups.

```python
# set the figure size
plt.figure(figsize=[15, 35])

# Calculate the linkage distances
mergings = linkage(q_data, method='ward') 

# cluster threshold - change to adjust the distance used to determine the number of clusters 
cluster_threshold = 15 

# Plot the dendrogram
dendrogram(mergings,
           labels=q_data.index,
           orientation='right',
           color_threshold=cluster_threshold, 
           leaf_rotation=0,
           leaf_font_size=10 )

plt.vlines(cluster_threshold, 0, 3000, 'orange', linestyles='dashed')
print('Fig.2.5 \n')
plt.title('Hierarchical Clustering Dendogram illustrating the linkage analysis performed on the subject\'s answers')
plt.ylabel('subject Id')
plt.xlabel('Distance between clusters')
plt.show()
```

#### Generation of the cluster labels for each subject_Id 

```python
# Use fcluster to extract labels fro the selected cluster threshold
labels = fcluster(mergings, cluster_threshold, criterion='distance')

# get the subject_Id's from the original data
subject_Id =q_data.index 
```

```python
# Create a new dataframe to store the results in
clusters = pd.DataFrame({'Cluster' : labels}, index=subject_Id)
```

```python
# create a version of this dataframe sorted by cluster number
clusters_sorted = clusters.sort_values('Cluster')
# how many subjects are there in each cluster?
print('Table 2.2 The number of subjects in each cluster\n')
clusters_sorted['Cluster'].value_counts(sort=False)
```

```python
# check the smallest cluster to see if it matches the dendogram subject_Id's
print('Table 2.3 Check the labels from fcluster to check they match those in the dendogram\n')
clusters_sorted[clusters_sorted['Cluster'] == 4]
```

```python
# export the results to a csv
clusters.to_csv('./H_cluster_results.csv')
km_df.to_csv('./KMeans_cluster_results.csv')
```
