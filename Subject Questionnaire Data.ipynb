## Data Analysis
### Principle Component Analysis (PCA)  
PCA can be used to illustrate how many questions in the data set are required to explain the variance in the students answers. Can the data set be explained by using a smaller subset of the questions?

```python
# import the PCA module
from sklearn.decomposition import PCA

# create and instance - without specifying the number of components required
pca = PCA()

# fit the data
pca.fit(q_data)

# create a dataframe to store the outputs of the PCA analysis
# set the index to the question numbers
q_data_output = pd.DataFrame(index=cols)

# collect the explained variance attribute
q_data_output['variance_ratio'] = pca.explained_variance_ratio_

# cumulatively sum the explained variance of each question
q_data_output['cum_var'] = q_data_output['variance_ratio'].cumsum(axis=0)
```

```python
# plot the data and add a line to indicate 95%
q_data_output['cum_var'].plot()
plt.hlines(0.95, q_data.index[0], q_data.index[-1], 'r', linestyles='dashed')
plt.hlines(0.5, q_data.index[0], q_data.index[-1],
           'orange', linestyles='dashed')
plt.ylabel('% of the explained variance')
plt.title('The cumulative number of questions required \nto explain the variance in the dataset')

print('Fig 2.0 \n')
plt.show()
```

```python
# how many questions explain 50% of the data?
pca2 = PCA(n_components=0.5)
pca2.fit(q_data)
print('The number of questions required to explain 50% of the variance in this data set: {}'.format(
    pca2.n_components_))
```

```python
# how many questions explain 95% of the data?
pca3 = PCA(n_components=0.95)
pca3.fit(q_data)
print('The number of questions required to explain 95% of the variance in this data set: {}'.format(
    pca3.n_components_))
```

#### Summary of the PCA Results  
From Fig.2.0 it can be seen that there is an initial steep rise with up to 5 of the questions explaining 50% of the data's overall variance. It however takes 26 questions to explain 95% of the variance in the data. There is no clear 'elbow' observed in the graph. In conclusion there is little scope for dimensionality reduction in this questionnaire dataset and nearly all of the questions are important and required to explain the results.

### Clustering  
With the assumption, that the questionnaire was originally used to classify the students into characteristic groups. Can these clusters be derived from the data without knowing the context and structure of the questions. If these student's answers can be clustered into meaningful groups; how does this unsupervised clustering compare to the original supervised clustering? 

#### K-Means clustering  
First we will try and use K-Means clustering algorithm, to visualise the multi dimensional data in a 2D space and see how the data clusters. We can use the K-Means attribute inertia to measure the quality of the cluster separation. A plot of cluster inertia vs the number of clusters can be used to determine an ideal number of clusters for the data.

```python
from sklearn.cluster import KMeans

# create a list of cluster numbers
clusters = list(range(1, 21))
# create a list to hold the inertia values
inertia = []
# loop through clusters and fit the data to each cluster size and collect the inertia data
for c in clusters:
    # create an instance
    kmn = KMeans(n_clusters=c)
    kmn.fit(q_data)
    inertia.append(kmn.inertia_)

df = pd.DataFrame({'inertia': inertia}, index=clusters)
df.plot(marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.title('The Reduction in K-Means inertia \nwith an increasing number of clusters')
plt.xlim(0, 20)
plt.xticks(clusters)
print('Fig 2.1 \n')
plt.show()
```

#### K-Means Inertia Results  
The graph in Fig 2.1 shows that, similar to the PCA analysis above, there is not a distinct 'elbow' in the graph and therefore it does not suggest an optimum number of clusters to use.

#### Plot the K-Means  
How does the data look when clustered with K-Means? Below we plot the data projected into 2 dimensions, with 8 clusters to visualise how the well the clusters separate. 

```python
# generate and 8 clusters
kmn = KMeans(n_clusters=8, random_state=42)
kmn.fit_predict(q_data)

# get the cluster labels
Cluster = kmn.labels_

# use PCA to convert the 32 questions into a 2D space to graph the clusters
pca = PCA(n_components=2)
X = pca.fit_transform(q_data)

# create a dataframe to hold the data to plot
km_df = pd.DataFrame(
    {'x': X[:, 0], 'y': X[:, 1], 'Cluster': Cluster}, index=q_data.index)
```

```python
# plot the clusters
plt.figure(figsize=[12, 7.5])
sns.scatterplot(data=km_df, x='x', y='y', hue='Cluster',
                palette='rainbow', legend='full', s=150)
#plt.scatter(km_df['x'], km_df['y'], c=km_df['Cluster'], cmap='rainbow', s=150)
plt.title('K-Means clustering with 206 students grouped into 8 clusters \nand projected onto a 2 dimensional representation of the questionnaire data set')

print('Fig 2.2 \n')
plt.show()
```

```python
# How many students are there in each cluster?
print('Table 2.1 The number of students in each of the 8 K-Means clusters\n ')
print(km_df['Cluster'].value_counts(sort=False))
```

#### Hierarchical Clustering with a dendogram  
Given that these types of student surveys typically try to group and sub group the students by their characteristics; a hierarchical clustering approach may produce a better way of understanding the data. To do this a linkage function can be used to measure the distance between the observations in the data set and form the clusters. Starting from the nearest two points these are then linked to form a single point and the distance to the next point evaluated. The process repeats until you have a tree like structure. However there are several algorithms for computing the linkages and the statistical literature does not provide any clear direction on which to use. First we will try them all and see which methods produce a realistic clustering of the data.

```python
from scipy.cluster.hierarchy import linkage, dendrogram

methods = ['median', 'single', 'average',
           'weighted', 'centroid', 'complete', 'ward']
fig = plt.figure(figsize=(15, 15))
for i, c in enumerate(methods):
    ax = fig.add_subplot(4, 2, i+1)
    mergings = linkage(q_data, method=c)
    dendrogram(mergings,
               labels=None,
               orientation='right',
               leaf_rotation=0)
    ax.set_title(c + " distance measure", color='k')
print('Fig. 2.4 The clustering produced by different distance methods\n ')
fig.tight_layout()
plt.show()
```

#### Hierarchical Clustering with a dendogram using the ward distance measure

From Fig 2.4 it can be seen that when either the ward or complete method is used to measure the distance between data points a converging dendogram is produced as we would expect from this type of survey. The ward method was selected as it appears to better generalise the data into even clusters. A detailed dendogram is produced below, using a distance of 15 the data can be clustered into 6 characteristic groups.

```python
# set the figure size
plt.figure(figsize=[15, 35])

# Calculate the linkage distances
mergings = linkage(q_data, method='ward')

# cluster threshold - change to adjust the distance used to determine the number of clusters
cluster_threshold = 15

# Plot the dendrogram
dendrogram(mergings,
           labels=q_data.index,
           orientation='right',
           color_threshold=cluster_threshold,
           leaf_rotation=0,
           leaf_font_size=10)

plt.vlines(cluster_threshold, 0, 3000, 'orange', linestyles='dashed')
print('Fig.2.5 \n')
plt.title('Hierarchical Clustering Dendogram illustrating the linkage analysis performed on the student\'s answers')
plt.ylabel('Student Id')
plt.xlabel('Distance between clusters')
plt.show()
```

### The description of the original questionnaire  
_from: Measuring Readiness for Career Management in Higher Education, G.Crust et al, unpublished_   

Each of 32 questionnaire items in version 1 was developed from the URICA readiness statements by a team of higher education careers advisers at the University of Plymouth.  The questionnaireâ€™s content validity was established by a review panel of higher education careers advisers in collaboration with a chartered psychologist.  Statements that appeared ambiguous, leading, double barrelled or negatively framed were identified and re-written

Version 1 of the questionnaire contains four subscales of eight items, each subscale representing one stage in the change process model: Precontemplation (items 1, 5, 11, 13, 23, 26, 29, 31); Contemplation (2, 4, 8, 12, 15, 19, 21, 24); Action (3, 7, 10, 14, 17, 20, 25, 30) and Maintenance (6, 9, 16, 18, 22, 27, 28, 32). Responses were recorded using a five point Likert scale, coded on the questionnaire between 1 (strongly disagree) and 5 (strongly agree), and recoded for data analysis between -2 (strongly disagree) and +2 (strongly agree). In interpreting an individual questionnaire, a score for each subscale were calculated as the sum of the eight corresponding item scores. This made available integer scores between -16 (i.e. 8 responses each scoring -2) and +16.
